### Accessing World Bank Data using Python
Alternatively, the datasets can be downloaded from the World Bank website.

```python
import matplotlib.pyplot as plt
import requests
import pandas as pd

#get the data % read into the file gd.xls 
#get the url by clicking download to get the dataset
wb_data_query = "http://api.worldbank.org/v2/en/indicator/gc.dod.totl.gd.zs?downloadformat=excel"
r = requests.get(wb_data_query)
with open('gd.xls','wb') as output:
    output.write(r.content)  
```

```python
#parse the data into a data frame
govt_debt = pd.read_excel('gd.xls',sheetname='Data', skiprows=3, index_col=1)
govt_debt.head(5)
```

```python
#take the desired values & plot it
govt_debt = govt_debt.transpose()
govt_debt.head(5)
```

```python
govt_debt = govt_debt[['AUS','AFG','FRA','USA']]
govt_debt = govt_debt[45:] #slicing the data from row 45 until the end of the table (year 2002 onwards)
govt_debt.plot(lw=1)
plt.show()
#missing lines on the plot means data not available for that particular country for that period of year
```
